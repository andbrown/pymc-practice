```python
import pymc as pm
import seaborn
```

```python

d0 = [20, 28, 24, 20, 23]                # observed demand samples

with pm.Model() as m:
    d = pm.Gamma('theta', 1, 1)          # prior distribution
    pm.Poisson('d0', d, observed = d0)   # likelihood
    samples = pm.sample(10000)           # draw samples from the posterior
```

```python
samples.posterior["theta"]
```

```python
    
# seaborn.distplot(samples.posterior['theta'], fit=stats.gamma)
seaborn.distplot(samples.posterior['theta'])
```
